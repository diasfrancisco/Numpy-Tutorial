```python
import numpy as np
import random
```

```python
# Create an array with random elements of size 20x20
arr1 = np.random.randint(0,5,(20,20))
print(arr1)
```

```python
# Add zeros to the border of the array
pad_arr = np.pad(arr1, pad_width=1, mode='constant', constant_values=0)
print(pad_arr)
```

```python
# Created a 3D array of size 20x20x20
arr2 = np.random.randint(0,5,(20,20,20))
print(arr2)
```

```python
# Pad the 3D array with zeros on the border
pad_arr2 = np.pad(arr2, pad_width=1, mode='constant', constant_values=0)
print(pad_arr2)
```

```python
# Created a 3D array of size 20x40x10
arr3 = np.random.randint(0,5,(20,40,10))
print(arr3)
```

```python
# Pad the 3D array with zeros on the border
pad_arr3 = np.pad(arr3, pad_width=1, mode='constant', constant_values=0)
print(pad_arr3)
```
